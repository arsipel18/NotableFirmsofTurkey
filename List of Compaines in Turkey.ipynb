```python
from bs4 import BeautifulSoup
import requests

url = 'https://en.wikipedia.org/wiki/List_of_companies_of_Turkey'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find_all('table', class_ = 'wikitable sortable')[1]
```

```python
table = soup.find_all('table', class_ = 'wikitable sortable')[1]
```

```python
table.find_all('th')
```

```python
table_titles = table.find_all('th')
word_table_titles = [title.text.strip() for title in table_titles]
print(word_table_titles)
```

```python
import pandas as pd
df = pd.DataFrame(columns=word_table_titles)
print(df)
```

```python
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    lenght = len(df)
    df.loc[lenght] = individual_row_data
```

```python
df
```

```python
df.to_csv(r'MyFilePath\List_of_companies_of_Turkey.csv', index = False)
```
